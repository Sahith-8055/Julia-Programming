```julia
x = 1:15
y = rand(15)
```

```julia
plot(x, y)
```

```julia
z = rand(15)
```

```julia
plot!(x, z)
```

```julia
p = plot(x, y)
plot!(p, x, z)
```

```julia
plot(x, y, title = "One Random Line Chart")
```

```julia
plot(x, y, title="Random Line Chart", label="Line Chart 1")
```

```julia
plot(x, y, xlabel="X Axis Label", ylabel="Y Axis Label", label="Line Chart 1", title="Random Line Chart")
```

```julia
scatter(x, y)
```

```julia
scatter(x, y, xlabel="X Axis Label", ylabel="Y Axis Label", title="Random Scatter Plot", label="Scatter Plot 1")
```

```julia
scatter!(x, z, label="Scatter Plot 2")
```

```julia
bar(x, y)
```

```julia
bar(x, y, xlabel="X Axis Label", ylabel="Y Axis Label", title="Random Bar Plot", label="Bar Plot 1")
```

```julia
bar!(x, z, label="Bar Plot 2")
```

```julia
histogram(y)
```

```julia
histogram(y, label="Histogram Plot 1", title="Histogram Plot", xlabel="X Axis Label", ylabel="Frequency")
```

```julia
histogram!(z, label="Histogram Plot 2")
```

```julia
p1 = plot(x, y, xlabel="X Axis Label", ylabel="Y Axis Label", label="Random Plot")
p2 = scatter(x, y, xlabel="X Axis Label", ylabel="Y Axis Label", label="Scatter Plot")
p3 = bar(x, y, xlabel="X Axis Label", ylabel="Y Axis Label", label="Bar Plot")
p4 = histogram(y, xlabel="X Axis Label", ylabel="Frequency", label="Histogram Plot")

plot(p1, p2, p3, p4, layout=(2, 2), legend=false)
```
